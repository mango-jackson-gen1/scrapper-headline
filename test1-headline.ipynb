### Part 1 - Set up 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36" 
}

response = requests.get("https://top.baidu.com/board?tab=realtime", headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')

response.status_code   # 200 means OK
```

### Part 2 - Identiy the loop structure

```python
headlines = doc.find_all("div", class_ ="category-wrap_iQLoo")   # 51 items
```

```python
#len(headlines)
```

```python
#headlines[1]
```

```python
rows = []
for headline in headlines:
    row = {}

    try:
        row['rank'] = headline.find(class_ = 'index_1Ew5p').text.strip()
    except:
        row['rank'] = None

    row ['title'] = headline.find(class_ ="c-single-text-ellipsis").text.strip()
    
    try: 
        row['subhead'] = headline.find(class_ ='hot-desc_1m_jR').text.strip()
    except:
        row['subhead'] = None

    try:
        row['count'] = headline.find(class_ = 'hot-index_1Bl1a').text.strip()
    except:
        row['count'] = None


    try:
        row['tag'] = headline.find(class_ = 'c-text').text.strip()
    except:
        row['tag'] = None

    try:
        row['link'] = headline.find('a')['href']
    except:
        row['link'] = None


    # try:
    #     row['premium'] = story.find(class_ = 'sr-only').text
    # except:
    #     row['premium'] = None
    
    


    # try:
    #     row['byline'] = story.find(class_ = 'article__byline').text.strip()
    # except:
    #     row['byline'] = None
    
    # img = story.find("img")
    # try:
    #    row['image_url'] = story.find('img')['src']
    # except:
    #    row['image'] = None

    # try:
    #     row['article_type'] = story.find(class_ = 'article__type').text
    # except:
    #     row['article_type'] = None





    rows.append(row)


len(rows)
```

### Part 3 - save as csv 

```python
import pandas as pd
df = pd.DataFrame(rows)
```

### Part 3.2 Saving method 1 - append to existing csv


```python
from datetime import datetime

# Add a new column for today's date

# You could just use datetime.now() and get the entire datetime
# or another strftime but this is fine
df['scrape_date'] = datetime.now().strftime("%Y-%m-%d_%H.%M")
df.head()
```

Check what already exists

```python
# If it exists, open it
# If it doesn't, just make a blank dataframe
# could also use os.path.exists to check if the file exists
# but honestly try/except is the easiest route to go here
try:
    existing_df = pd.read_csv("baidu-scrape1.csv")
except:
    existing_df = pd.DataFrame([])
existing_df.head()
```

```python
# Combine our new dataframe and our old dataframe
# ignore_index=True 
combined = pd.concat([df, existing_df], ignore_index=True)
combined.head()
```

```python
combined.to_csv("baidu-scrape1.csv", index=False)
```
